```python
import os
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Activation,Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2
import matplotlib.pyplot as plt
```

```python
epochs=15
width=height=224
```

```python
images=[]
ages=[]
for file_name in os.listdir('crop_part1')[0:9000]:
  parts=file_name.split('_')
  ages.append(int(parts[0]))
  image=cv2.imread(f'crop_part1/{file_name}')
  image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
  images.append(image)
```

```python
Images=pd.Series(images,name='Images')
Ages=pd.Series(ages,name='Ages')
df=pd.concat([Images,Ages],axis=1)
df.head()
```

```python
df = df[df['Ages'] < 90]
df = df[df['Ages'] >= 4]
df.head()
```

```python
X = []
Y = []

for i in range(len(df)):
    df['Images'].iloc[i] = cv2.resize(df['Images'].iloc[i], (width, height))
    X.append(df['Images'].iloc[i])
    Y.append(df['Ages'].iloc[i])

X = np.array(X)
Y = np.array(Y)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size = 0.2)
```

```python
data_generator = ImageDataGenerator(
                                  rescale = 1./ 255,
                                  horizontal_flip = True,
                                  )

train_data = data_generator.flow(X_train,
                                 Y_train,
                                 batch_size= 32,)

val_data = data_generator.flow(X_val,
                               Y_val,
                               batch_size= 32,)
```

```python
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(width, height, 3),
    weights = 'imagenet',
    pooling = 'avg',
    include_top = False,
    )
```

```python
for layer in base_model.layers[:-4]:
  layer.trainable = False
```

```python
model = tf.keras.models.Sequential([
    base_model,
    Dense(1, activation='relu')                          
])
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=0.001),loss = tf.keras.losses.mean_squared_error)
```

```python
model.fit(train_data,validation_data = val_data,epochs = epochs)
```
